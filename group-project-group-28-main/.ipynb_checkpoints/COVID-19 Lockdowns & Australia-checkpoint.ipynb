## Summary of Project
The following document contains a detailed analysis of COVID-19 cases and lockdown data in Australia. Due to the severity and international impact of COVID-19, our team has decided that analysing the types of lockdowns and their effects on COVID-19 cases would come up with useful and relevant results. This project details the effects of lockdowns on covid cases and how covid cases can be used to determine features of lockdowns.

#### Datasets
All the data used in this project contains records from **25/01/2020** up until **30/9/2021** as this was the date we downloaded the data. 

The first dataset that we will be using was obtained from M3IT’s [COVID-19_Data](https://github.com/M3IT/COVID-19_Data) repository on GitHub. It is a dataet detailing many COVID-19 case figures for each day and for each state. Daily case numbers, vaccines, deaths and tests are all detailed in this data.

The second dataset we use is the lockdown data which detailed when lockdowns occurred in each state and what gatherings were allowed or not. The types of gatherings include: community sport, gyms, schools, places of worship, clubs, dine in and indoor/outdoor gatherings (Records are in binary: 0 for gathering of at least 2, 1 for no gatherings). This data was created manually by Eric Tian using records from many news articles that reported on the restrictions of COVID-19 lockdowns in Australia.

#### Methods
This project contains the use of various techniques to express our analysis of COVID-19 cases and lockdown data.
- **K-means clustering:**
    - Visualisation of the map of Australia and it's COVID-19 cases
- **K-nearest neighbour:**
    - Predict if the state would be in lockdown or not depending on the number of cases
- **Linear Regression:**
    - Predict the length of the lockdown 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.feature_selection import RFE
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score, KFold
from sklearn.model_selection import GridSearchCV

%matplotlib inline
```

```python
# Read COVID-19 lockdown data
covid_lockdown = pd.read_csv("data/Lockdown.csv")
covid_lockdown.tail(8)
```

```python
# Read COVID-19 data
covid = pd.read_csv("data/COVID_AU_state.csv")
covid.tail(8)
```

# Preparing and Cleaning the data

We will delete unnecessary columns that aren't related to the questions we are answering

```python
# Checking for null values
covid.isnull().sum()
```

```python
# Removing null values by deleting the row.
covid.drop(4920, axis=0, inplace=True)
covid.tail()
```

# Joining the two data sets
We joined the lockdown and the covid cases dataset

```python
covid['state']=covid['state'].astype('string')
covid_lockdown['state']=covid_lockdown['state'].astype('string')
```

```python
covid_case_lockdown = pd.merge(covid, covid_lockdown)
covid_case_lockdown.head(8)
```

# Linear regression

```python
train, test = train_test_split(covid_case_lockdown, test_size=.2, random_state = 420)
```

```python
X_train = train[['Schools', 'Community Sport', 'Clubs', 'Gyms', 'Dine in', 'Place worship', 'Indoor gathering', 'Outdoor gathering']]
Y_train = train['confirmed']
X_test = test[['Schools', 'Community Sport', 'Clubs', 'Gyms', 'Dine in', 'Place worship', 'Indoor gathering', 'Outdoor gathering']]
Y_test = test['confirmed']
```

```python
reg = linear_model.LinearRegression()
reg.fit(X_train, Y_train)
```

```python
train_preds = reg.predict(X_train)
print("MSE:", mean_squared_error(Y_train, train_preds))
print("R Squared:", r2_score(Y_train, train_preds))
```

```python
test_preds = reg.predict(X_test)
print("MSE:", mean_squared_error(Y_test, test_preds))
print("R Squared:", r2_score(Y_test, test_preds))
```

# Interpretation
The R squared values for both train and test are quite low with  17% and 18.9% respectivly. The meansquared errors are quite high

# Logistic regression by state

```python
states = (covid_case_lockdown.loc[covid_case_lockdown['state_abbrev'] == 'ACT'],
        covid_case_lockdown.loc[covid_case_lockdown['state_abbrev'] == 'NSW'],
        covid_case_lockdown.loc[covid_case_lockdown['state_abbrev'] == 'NT'], 
        covid_case_lockdown.loc[covid_case_lockdown['state_abbrev'] == 'QLD'], 
        covid_case_lockdown.loc[covid_case_lockdown['state_abbrev'] == 'SA'], 
        covid_case_lockdown.loc[covid_case_lockdown['state_abbrev'] == 'TAS'], 
        covid_case_lockdown.loc[covid_case_lockdown['state_abbrev'] == 'VIC'], 
        covid_case_lockdown.loc[covid_case_lockdown['state_abbrev'] == 'WA'])
```

```python
# Predicting the lockdown with Logistic Regression
# Splitting the data into training(80%) and testing data (20%)
def train_and_test(data):
    X = data[['Schools', 'Community Sport', 'Clubs', 'Gyms', 'Dine in', 'Place worship', 'Indoor gathering', 'Outdoor gathering']]
    Y = data['confirmed']
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=420)
    return X_train, X_test, Y_train, Y_test
```

```python
train_test_states = []
for i in states:
    train_test_states.append(train_and_test(i))
```

```python
def log_regression_state(train_test_state, lr):
    lr.fit(train_test_state[0], train_test_state[2])
    Y_hat_train_state = lr.predict(train_test_state[0])
    Y_hat_test_state = lr.predict(train_test_state[1])
    return Y_hat_train_state, Y_hat_test_state
```

```python
lr = linear_model.LogisticRegression(max_iter=140)
y_hat_state = []
for i in range(0, 8):
    y_hat_state.append(log_regression_state(train_test_states[i], lr))
    
train_lr_accuracy_scores, test_lr_accuracy_scores, f1_scores, recall_scores, precision_scores=([] for i in range(5))

for i in range(0, 8):
    train_lr_accuracy_scores.append(accuracy_score(y_hat_state[i][0], train_test_states[i][2]))
    test_lr_accuracy_scores.append(accuracy_score(y_hat_state[i][1], train_test_states[i][3]))
    
lr_eval = pd.DataFrame({'Train_accuracy':train_lr_accuracy_scores, 'Test_accuracy':test_lr_accuracy_scores,})
lr_eval.index=['ACT', 'NSW', 'NT', 'QLD', 'SA', 'TAS', 'VIC', 'WA']

lr_eval
```

# Logistic regression country

```python
# Predicting the lockdown with Logistic Regression
# Splitting the data into training(80%) and testing data (20%)
train, test = train_test_split(covid_case_lockdown, test_size=0.2, random_state=420)
print(train.shape)
print(test.shape)
```

```python
# # Build your Logistic Regression model
X_train = train[['Schools', 'Community Sport', 'Clubs', 'Gyms', 'Dine in', 'Place worship', 'Indoor gathering', 'Outdoor gathering']]
Y_train = train['confirmed']
X_test = test[['Schools', 'Community Sport', 'Clubs', 'Gyms', 'Dine in', 'Place worship', 'Indoor gathering', 'Outdoor gathering']]
Y_test = test['confirmed']
lr = LogisticRegression(max_iter=140)
lr.fit(X_train, Y_train)
```

```python
# Do predictions on test and train set
train_preds = lr.predict(X_train)
test_preds = lr.predict(X_test)
```

```python
print("Train accuracy: ", (accuracy_score(Y_train, train_preds)))
print("Test accuracy: ", (accuracy_score(Y_test, test_preds)))
```

# Interpretation
Accuracy scores are too low to be considered with train model at 0.548 and test model at 0.515. There seems to be no overfitting in this model as the train and test scores are quite similar.

# RFE model country

```python
train_accuracies= []
test_accuracies= []
for i in range(X_train.shape[1]):
    lr = LogisticRegression(max_iter=140)
    rfe = RFE(estimator = lr, n_features_to_select=i+1, step = 1)
    rfe.fit(X_train, Y_train)
    
    train_rfe_preds = rfe.predict(X_train)
    test_rfe_preds = rfe.predict(X_test)

    train_accuracies.append(accuracy_score(Y_train, train_rfe_preds))
    test_accuracies.append(accuracy_score(Y_test, test_rfe_preds))
```

```python
#Fitting the accuracy scores on a line graph
plt.figure(figsize=(12,8))
plt.plot(range(X_train.shape[1]), train_accuracies)
plt.plot(range(X_train.shape[1]), test_accuracies)
```

# Interpretation
The more features used in this model the less accurate with more features, but the accuracy score it reduces is negligible. The difference in accuracy score was only around 0.005 approximately meaning RFE would be meaningless.

# KNN Model

```python
# Trying possible number of K-neighbors from 1 with a step of 5 capped until 70 neighbors
parameters_grid = {'n_neighbors': range(1, 70, 4)}
knn_clf = KNeighborsClassifier()
gs_knn = GridSearchCV(knn_clf, parameters_grid, scoring = 'accuracy', cv = KFold(n_splits = 4, shuffle = True))
```

```python
# Fit knn model with training data for each states
k_values=[]
accuracy=[]
for i in range(0,8):
    gs_knn.fit(train_test_states[i][0], train_test_states[i][2])
    k_values.append(gs_knn.best_params_['n_neighbors'])
    accuracy.append(gs_knn.best_score_)
```

```python
print("Best K value (ACT):", k_values[0])
print("Accuracy (ACT):", accuracy[0])

print("Best K value (NSW):", k_values[1])
print("Accuracy (NSW):", accuracy[1])

print("Best K value (NT):", k_values[2])
print("Accuracy (NT):", accuracy[2])

print("Best K value (QLD):", k_values[3])
print("Accuracy (QLD):", accuracy[3])

print("Best K value (SA):", k_values[4])
print("Accuracy (SA):", accuracy[4])

print("Best K value (TAS):", k_values[5])
print("Accuracy (TAS):", accuracy[5])

print("Best K value (VIC):", k_values[6])
print("Accuracy (VIC):", accuracy[6])

print("Best K value (WA):", k_values[7])
print("Accuracy (WA):", accuracy[7])
```

Then we build the knn model for each states based on the gridsearch cv results

```python
# Building KNN classifier model based on best K value
knn_ACT = KNeighborsClassifier(n_neighbors=5)
knn_NSW = KNeighborsClassifier(n_neighbors=9)
knn_NT = KNeighborsClassifier(n_neighbors=9)
knn_QLD = KNeighborsClassifier(n_neighbors=49)
knn_SA = KNeighborsClassifier(n_neighbors=13)
knn_TAS = KNeighborsClassifier(n_neighbors=5)
knn_VIC = KNeighborsClassifier(n_neighbors=17)
knn_WA = KNeighborsClassifier(n_neighbors=9)

knn_models=[knn_ACT, knn_NSW, knn_NT, knn_QLD, knn_SA, knn_TAS, knn_VIC, knn_WA ]
```

```python
# Fitting the model
def fit_knn(train_test_state, clf):
    clf.fit(train_test_state[0], train_test_state[2])
    y_hat = clf.predict(train_test_state[1])
    accuracy = accuracy_score(y_hat, train_test_state[3])
    return accuracy
```

```python
knn_metrics=[]
for i in range(0,8):
    knn_metrics.append(fit_knn(train_test_states[i], knn_models[i]))
col_names=['Accuracy']
knn_eval=pd.DataFrame(data=knn_metrics, columns=col_names)
knn_eval.index=['ACT', 'NSW', 'NT', 'QLD', 'SA', 'TAS', 'VIC', 'WA']
knn_eval
```
